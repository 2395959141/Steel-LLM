### 检查CCIC3-HQ数据集的处理

```python
# 使用jsonlines库加载jsonl文件
import jsonlines

# 指定数据集路径
dataset_path = "/home/ytllm/.cache/modelscope/datasets/BAAI/CCI3-HQ/data/part_000000.jsonl"

# 打开jsonl文件并读取内容
with jsonlines.open(dataset_path) as reader:
    # 打印文件中的前5条记录作为示例
    print("JSONL文件前5条记录:")
    for i, item in enumerate(reader):
        print(f"记录 {i+1}: {item}")
        if i >= 4:  # 只打印前5条
            break
            
    # 获取总记录数(需要重新读取文件)
    total_records = sum(1 for _ in jsonlines.open(dataset_path))
    print(f"\n文件总记录数: {total_records}")
    # 重新打开文件读取第一条记录的结构
    with jsonlines.open(dataset_path) as reader:
        first_record = next(iter(reader))
        
    # 打印记录中的字段信息
    print("\n记录字段结构:")
    if isinstance(first_record, dict):
        print("字段数量:", len(first_record))
        print("字段名称和类型:")
        for key, value in first_record.items():
            print(f"- {key}: {type(value).__name__}")
    else:
        print("记录不是字典类型，实际类型为:", type(first_record).__name__)

```

```python
from data.pretrain_data_prepare.step1_data_process import *
# 创建CCI3_HQFormatHandler实例来处理数据
handler = CCI3_HQFormatHandler(dataset_path, "demo/CCI3-HQ", "CCI3-HQ")

# 打印数据集处理信息
print("\n数据集处理信息:")
print(f"数据集名称: {handler.dataset_name}")
print(f"输入路径: {handler.input_path}")

# 初始化输出文件路径
output_path = "demo/CCI3-HQ/processed_data.jsonl"
os.makedirs(os.path.dirname(output_path), exist_ok=True)

# 打开输出文件准备写入
with open(output_path, "w", encoding="utf-8") as fout:
    # 读取并处理数据
    with jsonlines.open(dataset_path) as reader:
        processed_count = 0
        for i, item in enumerate(reader):
            # 处理每条记录
            success = handler.process_one_line(json.dumps(item), fout)
            if success:
                processed_count += 1
            if i >= 4:  # 只处理前5条
                break

# 读取并显示处理后的结果
print("\n处理后的前5条记录:")
with open(output_path, "r", encoding="utf-8") as fin:
    for i, line in enumerate(fin):
        print(f"记录 {i+1}:")
        print(json.loads(line))
        print()
        if i >= 4:  # 只显示前5条
            break
```

### 测试chinese-fine-web-edu

```python
# 导入必要的库
import pyarrow.parquet as pq
import os

dataset_path = "/home/ytllm/.cache/modelscope/datasets/opencsg/chinese-fineweb-edu/data/00000.parquet"

# 创建ChineseFineWebEduFormatHandler实例来处理数据
handler = ChineseFineWebEduFormatHandler(dataset_path, "demo/chinese-fine-web-edu", "chinese-fine-web-edu")

# 打印数据集处理信息
print("\n数据集处理信息:")
print(f"数据集名称: {handler.dataset_name}")
print(f"输入路径: {handler.input_path}")

# 初始化输出文件路径
output_path = "demo/chinese-fine-web-edu/processed_data.jsonl"
os.makedirs(os.path.dirname(output_path), exist_ok=True)

# 打开输出文件准备写入
with open(output_path, "w", encoding="utf-8") as fout:
    # 读取并处理数据
    table = pq.read_table(dataset_path)  # 读取parquet文件
    processed_count = 0
    for i in range(min(5, len(table))):  # 只处理前5条
        # 将每条记录转换为字典格式
        item = {col: table[col][i].as_py() for col in table.column_names}
        # 处理每条记录
        success = handler.process_one_line(json.dumps(item), fout)
        if success:
            processed_count += 1

# 读取并显示处理后的结果
print("\n处理后的前5条记录:")
with open(output_path, "r", encoding="utf-8") as fin:
    for i, line in enumerate(fin):
        print(f"记录 {i+1}:")
        print(json.loads(line))
        print()
        if i >= 4:  # 只显示前5条
            break
```
